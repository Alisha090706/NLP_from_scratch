```python
sentence = "Eiffel Tower was built in 1889. It is located in Paris. It is one of the most famous landmarks in the world. Many tourists visit it every year."
```

```python
import nltk
words = nltk.word_tokenize(sentence)
```

```python
tag_elements = nltk.pos_tag(words)
```

```python
nltk.download('maxent_ne_chunker_tab')
```

```python
nltk.download('words')
```

```python
nltk.ne_chunk(tag_elements).draw()
```
