```python
words = ['writing', 'writes', 'written', 'writer', 'wrote', 'eating', 'eaten', 'eats', 'ate', 'runner', 'running', 'ran']
```

```python
import nltk

nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
'''POS Tags: 
a: Adjective
ad: Adverb
n: Noun
v: Verb
By default, it is set to 'n' (Noun)'''
for word in words:
    print(f"{word} --> Noun: {lemmatizer.lemmatize(word)}, Verb: {lemmatizer.lemmatize(word, 'v')}")
```

Lemmatization is better than stemming but it takes more time to process words.
