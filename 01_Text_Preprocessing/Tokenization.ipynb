```python
corpus = '''Natural language processing (NLP) is a subfield of artificial intelligence (AI) that focuses on the interaction between computers and humans through natural language. The ultimate goal of NLP is to enable computers to understand, interpret, and generate human language in a way that is valuable.'''
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
### Tokenization 
## Paragraphs --> Sentences --> Words
from nltk.tokenize import sent_tokenize
documents = sent_tokenize(corpus)
```

```python
from nltk.tokenize import word_tokenize

for sentence in documents:
    words = word_tokenize(sentence)
    print(words)
```

```python
from nltk.tokenize import wordpunct_tokenize

for sentence in documents:
    words = wordpunct_tokenize(sentence)
    print(words)
```

The only difference between `word_tokenize` and `wordpunct_tokenize` is that `word_tokenize` treats most punctuations separately but keeps contractions (don't, wasn't etc) whereas `wordpunct_tokenize` splits all punctuations

```python
from nltk.tokenize import TreebankWordTokenizer

tokenizer = TreebankWordTokenizer()
for sentence in documents:
    words = tokenizer.tokenize(sentence)
    print(words)
    
```
